```python
import pandas as pd

# Загружаем данные
s_df_mk = pd.read_csv('Desktop/data/samokat_makarony.csv')
s_df_mk['price'] = s_df_mk['price'].astype(str).str.replace('₽', '').str.strip()
# Список известных брендов (дополните его актуальными брендами для вашей категории)
BRANDS = ['Barilla', 'Foody', 'Grand di pasta']

# Функция для извлечения бренда из названия
def extract_brand(name):
    name = str(name).lower()
    for brand in BRANDS:
        if brand.lower() in name:
            return brand
    return None

# Применяем функцию к столбцу name и заполняем brand
s_df_mk['brand'] = s_df_mk['name'].apply(extract_brand)

# Удаляем строки с пустыми брендами
s_df_mk = s_df_mk.dropna(subset=['brand'])

# Сохраняем результат
s_df_mk.to_csv('samokat_makarony_filtered.csv', index=False)
```
