```python
import pandas as pd
 """Модуль для очистки данных Вкусвилла и EDA. Удаляем бренды, которых нет в Самокате"""

v_df_m = pd.read_csv("Desktop/vkusvill_myaso.csv")
MYASO_BRANDS = ['Ближние Горки', 'Мираторг', 'ИНДИ ЛАЙТ', 'Троекурово', 
    'Наша Птичка', 'Рококо']
# Удаляем всё после "руб" включая пробелы и спецсимволы
v_df_m['price'] = v_df_m['price'].str.replace(r'руб.*', '', regex=True).str.strip()
v_df_m['price'] = pd.to_numeric(v_df_m['price'], errors='coerce')

# Проверка результата
print(v_df_m['price'].head())
filtered_meat = v_df_m[v_df_m['brand'].isin(MYASO_BRANDS)]
filtered_meat.to_csv('vkusvill_myaso_filtered.csv', index=False)


v_df_mk = pd.read_csv("Desktop/vkusvill_makarony.csv")
MAK_BRANDS = ["Barilla", "Foody", "Grand Di Pasta"]


v_df_mk['price'] = v_df_mk['price'].str.replace(r'руб.*', '', regex=True).str.strip()
v_df_mk['price'] = pd.to_numeric(v_df_m['price'], errors='coerce')

# Проверка результата
print(v_df_mk['price'].head())
filtered_pasta = v_df_mk[v_df_mk['brand'].isin(MAK_BRANDS)]
filtered_pasta.to_csv('vkusvill_makarony_filtered.csv', index=False)


v_df_s = pd.read_csv("Desktop/vkusvill_syry.csv")
SYR_BRANDS = ["Bonfesto", "Dolce Granto", "Galbani", "Hochland", "Laplandia", "Mlekara Sabac", "Natura",
             "Pretto", "Unagrande", "Viola", "Ламбер", "ЭКОМИЛК", "ЭкоНива"]  # <- Исправлено: "Barilla" вместо "Вarilla"


v_df_s['price'] = v_df_s['price'].str.replace(r'руб.*', '', regex=True).str.strip()
v_df_s['price'] = pd.to_numeric(v_df_s['price'], errors='coerce')

# Проверка результата
print(v_df_s['price'].head())
filtered_syry = v_df_s[v_df_s['brand'].isin(SYR_BRANDS)]
filtered_syry.to_csv('vkusvill_syry_filtered.csv', index=False)


```

```python
def eda(df):
    print("Информация о таблице:")
    print(df.info())

    print(" Описательная статистика:")
    print(df.describe())
    
    print("Количество пропущенных значений:")
    print(df.isnull().sum())
    
    print("Количество уникальных значений в каждом столбце:")
    print(df.nunique())
    
    print("Первые 5 строк датафрейма:")
    print(df.head())
eda(v_df_m)
eda(v_df_mk)
eda(v_df_s)    
```
